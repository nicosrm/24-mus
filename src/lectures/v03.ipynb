```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
x0 = 3

x_plot = np.linspace(0, 6, 1000)

plt.plot(x_plot, norm.pdf(x_plot, 2, np.sqrt(0.5)), color='r')
plt.plot(x_plot, norm.pdf(x_plot, 4, np.sqrt(0.5)), color='b')

plt.axis([0, 6, 0, 0.6])

plt.axvline(x=x0, color='black', linestyle='--', linewidth=1.5)
plt.annotate(f'$x_0 = {x0}$', xy=(x0,0.6))

plt.hlines(y=0, xmin=0, xmax=x0, color='red', linewidth=10)
plt.hlines(y=0, xmin=x0, xmax=6, color='blue', linewidth=10)

red_section = np.linspace(x0, 6)
plt.fill_between(red_section, norm.pdf(red_section, 2, np.sqrt(0.5)), color='lightcoral')

blue_section = np.linspace(0, x0)
plt.fill_between(blue_section, norm.pdf(blue_section, 4, np.sqrt(0.5)), color='lightblue')

plt.show()
```

```python
x0 = 3

x_plot = np.linspace(0, 6, 1000)

plt.plot(x_plot, norm.pdf(x_plot, 2, np.sqrt(0.5)), color='salmon', label='$C_1$')
plt.plot(x_plot, norm.pdf(x_plot, 4, np.sqrt(0.5)), color='lightblue', label='$C_2$')

plt.plot(x_plot, 0.9 * norm.pdf(x_plot, 2, np.sqrt(0.5)), color='r', label=r'$0.9 \cdot C_1$')
plt.plot(x_plot, 0.1 * norm.pdf(x_plot, 4, np.sqrt(0.5)), color='b', label=r'$0.1 \cdot C_2$')

plt.axis([0, 6, 0, 0.6])

plt.axvline(x=x0, color='gray', linestyle='--', linewidth=1.5)
plt.annotate(f'$x_0$', xy=(x0,0.6))

plt.axvline(x=3.53, color='black', linestyle='--', linewidth=1.5)
plt.annotate(f'$x_0^*$', xy=(3.54,0.6))

plt.legend()
plt.show()
```
